```python
def convert_duration_to_seconds(duration_str):
    try:
        digits = ''.join(filter(str.isdigit, duration_str))
        if 'hour' in duration_str or 'hr' in duration_str:
            return int(digits) * 3600
        elif 'minute' in duration_str or 'min' in duration_str:
            return int(digits) * 60
        elif 'second' in duration_str or 'sec' in duration_str:
            return int(digits)
        else:
            return None
    except:
        return None
ufo_df['duration (seconds)'] = ufo_df['duration (hours/min)'].apply(convert_duration_to_seconds)
```

```python
from textblob import TextBlob
from collections import Counter
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from wordcloud import WordCloud

selected_columns = [
    'comments'
]


ufo_df.dropna(subset=['comments'], inplace=True)
ufo_df['comments'] = ufo_df['comments'].astype(str)
ufo_df['sentiment'] = ufo_df['comments'].apply(lambda x: TextBlob(x).sentiment.polarity)

ufo_df['year'] = pd.to_datetime(ufo_df['datetime']).dt.year
avg_sentiment_by_year = ufo_df.groupby('year')['sentiment'].mean()

plt.figure(figsize=(12, 6))
avg_sentiment_by_year.plot(kind='line', marker='o', color='skyblue')
plt.title('Average Sentiment of UFO Sightings Comments Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Sentiment')
plt.grid(True)
plt.show()

stop_words = set(stopwords.words('english'))
ufo_df['tokenized_comments'] = ufo_df['comments'].apply(lambda x: nltk.word_tokenize(x.lower()))
ufo_df['tokenized_comments'] = ufo_df['tokenized_comments'].apply(lambda x: [word for word in x if word.isalnum() and word not in stop_words])

all_words = [word for comment in ufo_df['tokenized_comments'] for word in comment]
word_freq = Counter(all_words)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Common Keywords in UFO Sightings Comments')
plt.axis('off')
plt.show()
```

```python
grouped_data = ufo_df.groupby(['year', 'country'])['datetime'].count().reset_index(name='total_sightings')
grouped_data['total_population'] = ufo_df.groupby(['year', 'country'])['country population'].max().values
grouped_data['sightings_per_capita'] = grouped_data['total_sightings'] / grouped_data['total_population']

plt.figure(figsize=(12, 8))
for country, group in grouped_data.groupby('country'):
    plt.plot(group['year'], group['sightings_per_capita'], label=country.upper(), lw=2)

plt.title('Normalized UFO Sightings Per Capita by Country Over Time (All Years)')
plt.xlabel('Year')
plt.ylabel('Sightings Per Capita')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
